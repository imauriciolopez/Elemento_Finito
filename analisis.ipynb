```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr=[[5, 2, 3],
     [4, 5, 6],
     [7, 587, 9]]

print(np.linalg.det(arr))
print(np.linalg.inv(arr))
```

```python
arr=[[15, 29],
     [34, 58],
     [13, 21]]

print(np.linalg.pinv(arr))
print(arr @ np.linalg.pinv(arr) @ arr)
```
